```python
# Create a class Account with a private __balance and provide deposit(), withdraw(), and get_balance() methods.
class Account:
    def __init__(self, balance):
        self.__balance = balance
    def get_balance(self):
        return f"The available balance is {self.__balance}"
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"The balance after deposit is {self.__balance}"
        else:
            return "Invalid deposit amount!"
    def withdraw(self, with_amount):
        if with_amount <= self.__balance and with_amount > 0:
            self.__balance -= with_amount
            return f"The balance after withdrawal is {self.__balance}"
        else:
            return "Insufficient balance or invalid withdrawal amount!"
user = Account(5000)
print(user.get_balance())
dep_amount = float(input("Enter the amount to be deposited: "))
print(user.deposit(dep_amount))
with_amount = float(input("Enter the amount to be withdrawn: "))
print(user.withdraw(with_amount))
```

```python
# Create a class Student with private __name and __grade and use getter/setter methods to update values safely.
class Student:
    def __init__(self, name, grade):
        self.__name = name
        self.__grade = grade

    def get_name(self):
        return self.__name

    def set_name(self, new_name):
        self.__name = new_name

    def get_grade(self):
        return self.__grade

    def set_grade(self, new_grade):
        self.__grade = new_grade


name = input("Enter student name: ")
grade = int(input("Enter student grade: "))

student = Student(name, grade)

print("Student Name:", student.get_name())
print("Student Grade:", student.get_grade())

new_name = input("Enter new name to update: ")
student.set_name(new_name)

new_grade = int(input("Enter new grade to update: "))
student.set_grade(new_grade)

print("Updated Name:", student.get_name())
print("Updated Grade:", student.get_grade())
```

```python
# Create a class Car with private __speed and methods increase_speed() and decrease_speed().
class Car:
    def __init__(self, speed):
        self.__speed = speed

    def get_speed(self):
        return self.__speed

    def increase_speed(self, value):
        self.__speed = self.__speed + value

    def decrease_speed(self, value):
        self.__speed = self.__speed - value


speed = int(input("Enter starting speed: "))
car = Car(speed)
print("Current Speed:", car.get_speed())

inc = int(input("Enter value to increase speed: "))
car.increase_speed(inc)
print("Speed After Increase:", car.get_speed())

dec = int(input("Enter value to decrease speed: "))
car.decrease_speed(dec)
print("Speed After Decrease:", car.get_speed())
```

```python
# Create a class Employee with private __salary and method to safely update salary using validation.
class Employee:
    def __init__(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def update_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
        else:
            print("Invalid salary. Must be greater than 0.")


salary = int(input("Enter starting salary: "))

emp = Employee(salary)
print("Current Salary:", emp.get_salary())

new_salary = int(input("Enter new salary: "))
emp.update_salary(new_salary)
print("Updated Salary:", emp.get_salary())
```

```python
# Create a class Product with private __price and method apply_discount() without exposing price directly.
class Product:
    def __init__(self, price):
        self.__price = price

    def apply_discount(self, discount):
        if discount > 0 and discount < 100:
            self.__price = self.__price - (self.__price * discount / 100)
        else:
            print("Invalid discount")

    def get_price(self):
        return self.__price

price = int(input("Enter product price: "))
p = Product(price)

discount = int(input("Enter discount percentage: "))
p.apply_discount(discount)
print("Price After Discount:", p.get_price())
```

```python
# Create a class User with private __password and a method check_password() for verification.
class User:
    def __init__(self, password):
        self.__password = password

    def check_password(self, entered_password):
        if entered_password == self.__password:
            return True
        else:
            return False

password = input("Set your password: ")
user = User(password)

entered = input("Enter password to verify: ")
if user.check_password(entered):
    print("Password is correct")
else:
    print("Password is wrong")
```

```python
# Create a Shape class and child classes Square and Circle that calculate area differently.
class Shape:
    def area(self):
        return 0

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


s = Square(5)
c = Circle(4)

print("Square Area:", s.area())
print("Circle Area:", c.area())
```

```python
# 15. Create a base Laptop class and inherit GamingLaptop with extra GPU attribute.
class Laptop:
    def __init__(self, brand, ram):
        self.brand = brand
        self.ram = ram

    def show_info(self):
        return "Brand: " + self.brand + ", RAM: " + str(self.ram) + "GB"


class GamingLaptop(Laptop):
    def __init__(self, brand, ram, gpu):
        super().__init__(brand, ram)
        self.gpu = gpu

    def show_gaming_info(self):
        return self.show_info() + ", GPU: " + self.gpu


brand = input("Enter laptop brand: ")
ram = int(input("Enter laptop RAM (GB): "))
gpu = input("Enter GPU model: ")

g = GamingLaptop(brand, ram, gpu)

print(g.show_info())
print(g.show_gaming_info())
```

```python
# 16. Demonstrate the use of super() to call parent class constructor inside a child class.
class Parent:
    def __init__(self, name):
        self.name = name

    def show(self):
        return "Parent Name: " + self.name


class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)   # calling parent constructor
        self.age = age

    def show_child(self):
        return "Child Name: " + self.name + ", Age: " + str(self.age)


name = input("Enter name: ")
age = int(input("Enter age: "))

c = Child(name, age)

print(c.show())
print(c.show_child())
```

```python
# 22. Create a class Fraction and overload operators + and -.
class Fraction:
    def __init__(self, num, den):
        self.num = num
        self.den = den

    def __add__(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def __sub__(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def show(self):
        print(self.num, "/", self.den)


f1 = Fraction(1, 2)
f2 = Fraction(1, 3)

f3 = f1 + f2
f4 = f1 - f2

f3.show()
f4.show()
```

```python
# Create an abstract class Animal with abstract methods eat() and sleep() and implement them in Dog.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def eat(self):
        pass

    @abstractmethod
    def sleep(self):
        pass

class Dog(Animal):
    def eat(self):
        return "Dog is eating"

    def sleep(self):
        return "Dog is sleeping"

d = Dog()
print(d.eat())
print(d.sleep())
```

```python
# Create an abstract class Shape with abstract method perimeter() and derive Rectangle and Triangle.
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def perimeter(self):
        pass


class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def perimeter(self):
        return 2 * (self.l + self.w)


class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c


r = Rectangle(5, 3)
t = Triangle(3, 4, 5)

print("Rectangle Perimeter:", r.perimeter())
print("Triangle Perimeter:", t.perimeter())
```

```python
# Create an abstract class Database with connect(), read(), write() and implement MySQLDatabase.
from abc import ABC, abstractmethod

class Database(ABC):
    @abstractmethod
    def connect(self):
        pass

    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def write(self):
        pass


class MySQLDatabase(Database):
    def connect(self):
        return "Connected to MySQL Database"

    def read(self):
        return "Reading data from MySQL"

    def write(self):
        return "Writing data to MySQL"


db = MySQLDatabase()

print(db.connect())
print(db.read())
print(db.write())
```

```python
# Create an abstract MediaPlayer class with play(), pause(), stop() and derive AudioPlayer and VideoPlayer.
from abc import ABC, abstractmethod

class MediaPlayer(ABC):
    @abstractmethod
    def play(self):
        pass

    @abstractmethod
    def pause(self):
        pass

    @abstractmethod
    def stop(self):
        pass


class AudioPlayer(MediaPlayer):
    def play(self):
        return "Audio Playing"

    def pause(self):
        return "Audio Paused"

    def stop(self):
        return "Audio Stopped"


class VideoPlayer(MediaPlayer):
    def play(self):
        return "Video Playing"

    def pause(self):
        return "Video Paused"

    def stop(self):
        return "Video Stopped"


a = AudioPlayer()
v = VideoPlayer()

print(a.play())
print(a.pause())
print(a.stop())

print(v.play())
print(v.pause())
print(v.stop())
```

```python
# Create an abstract FileHandler class with open(), read(), close() and implement TextFileHandler.
from abc import ABC, abstractmethod

class FileHandler(ABC):
    @abstractmethod
    def open(self):
        pass

    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def close(self):
        pass


class TextFileHandler(FileHandler):
    def open(self):
        return "Text file opened"

    def read(self):
        return "Reading text file"

    def close(self):
        return "Text file closed"


f = TextFileHandler()

print(f.open())
print(f.read())
print(f.close())
```
