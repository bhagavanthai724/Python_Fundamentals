```python
# SyntaxError – missing colon in if statement
if True
    print("hi")
```

```python
# IndentationError – code block not indented
def load():
print("fail")
```

```python
# TabError – mixed tabs and spaces in indentation
if True:
	print("ok")
    print("bad")
```

```python
# NameError – variable used before definition
print(model)
```

```python
# TypeError – adding string + integer
x = "10"
y = 2
print(x + y)
```

```python
# ValueError – cannot convert text to int
int("hello")
```

```python
# AttributeError – 'int' has no attribute append
a = 10
a.append(5)
```

```python
# KeyError – dictionary key does not exist
data = {"a": 1}
data["b"]
```

```python
# IndexError – list index out of range
arr = [1,2,3]
arr[10]
```

```python
# FileNotFoundError – file path does not exist
open("missing.csv")
```

```python
# PermissionError – writing to a protected directory
with open("/root/secret.txt", "w") as f:
    f.write("x")
```

```python
# ModuleNotFoundError – importing a non-existing package
import fake_llm_pkg
```

```python
# ImportError – function name incorrect
from math import sqrtt
```

```python
# OSError – invalid model path in HuggingFace load
from transformers import AutoModel
AutoModel.from_pretrained("wrong/path")
```

```python
# RuntimeError – manually raised runtime crash
raise RuntimeError("fail")
```

```python
# RuntimeError – tensor shape mismatch for matmul
import torch
a = torch.randn(3,4)
b = torch.randn(5,6)
a @ b
```

```python
# CUDAOutOfMemoryError – tensor too large for GPU memory
import torch
torch.randn((50000,50000), device="cuda")
```

```python
# CUDADeviceSideError – label out of range for class count
import torch, torch.nn as nn
logits = torch.randn(3,2)
labels = torch.tensor([0,2,5])
nn.CrossEntropyLoss()(logits, labels)
```

```python
# LinAlgError – matrix is singular (non-invertible)
import numpy as np
m = np.array([[1,2],[2,4]])
np.linalg.inv(m)
```

```python
# AxisError – axis does not exist in array
import numpy as np
x = np.array([1,2,3])
x.sum(axis=1)
```

```python
# ParserError – malformed CSV format
import pandas as pd
from io import StringIO
bad = "name,age\njohn,20\njane,abc,extra"
pd.read_csv(StringIO(bad))
```

```python
# EmptyDataError – reading empty CSV file
import pandas as pd
from io import StringIO
pd.read_csv(StringIO(""))
```

```python
# JSONDecodeError – invalid JSON structure
import json
json.loads("<html>Error</html>")
```

```python
# UnicodeDecodeError – wrong file encoding
open("non_utf8.txt", encoding="utf-8").read()
```

```python
# ConnectionError – server unreachable
import requests
requests.get("http://localhost:9999/no_server")
```

```python
# TimeoutError – request exceeded allowed time
import requests
requests.get("https://httpbin.org/delay/10", timeout=1)
```

```python
# TokenizationError – invalid input passed to tokenizer
from transformers import AutoTokenizer
tok = AutoTokenizer.from_pretrained("gpt2")
tok(None)
```

```python
# OSError – tokenizer folder missing/corrupt
from transformers import AutoTokenizer
AutoTokenizer.from_pretrained("broken/tokenizer")
```

```python
# ResponseValidationError – invalid type returned in schema
from pydantic import BaseModel
class Out(BaseModel):
    msg: str
Out(msg=123)
```

```python
# BrokenProcessPool – multiprocessing worker crashed
from concurrent.futures import ProcessPoolExecutor

def crash(x):
    raise ValueError("fail")
with ProcessPoolExecutor() as ex:
    list(ex.map(crash, [1,2,3]))
```
